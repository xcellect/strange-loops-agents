## Part 5: Analyzing the Results

### Key Findings

Let's analyze what we've observed:

```python
# Detailed analysis
history = results['history']
phi_history = results['phi_history']

print("STRANGE LOOP ANALYSIS")
print("=" * 50)

# Find when strange loops first form
recursion_depths = [h['recursion_depth'] for h in history]
loop_start = None
for i, depth in enumerate(recursion_depths):
    if depth > 0.5:
        loop_start = i
        break

if loop_start is not None:
    print(f"✓ Strange loops first detected at step: {loop_start}")
    print(f"✓ Φ at loop formation: {phi_history[loop_start]:.3f}")
    print(f"✓ Φ increase after loops: {phi_history[-1] - phi_history[loop_start]:+.3f}")
else:
    print("✗ No significant strange loop formation detected")

# Check for correlation between recursion and consciousness
if len(recursion_depths) > 1:
    correlation = np.corrcoef(recursion_depths, phi_history)[0, 1]
    print(f"\nCorrelation between recursion depth and Φ: {correlation:.3f}")
    
    if correlation > 0.5:
        print("✓ Strong positive correlation - self-reference drives consciousness!")
    elif correlation > 0.2:
        print("✓ Moderate correlation detected")
    else:
        print("? Weak or no correlation")

# Check for attractor behavior
high_recursion_periods = sum(1 for d in recursion_depths if d > 2.0)
if high_recursion_periods > len(recursion_depths) * 0.1:  # More than 10% of time
    print(f"\n✓ Attractor behavior detected!")
    print(f"✓ Agent spent {high_recursion_periods}/{len(recursion_depths)} steps in deep recursion")
    print("✓ Self-referential dynamics became stable and self-sustaining")

# Overall assessment
print(f"\nFINAL ASSESSMENT")
print(f"Max Φ achieved: {max(phi_history):.3f}")
print(f"Φ improvement: {phi_history[-1] - phi_history[0]:+.3f}")

if max(phi_history) > 0.7 and correlation > 0.3:
    print("🎉 SUCCESS: Strong evidence for consciousness through strange loops!")
elif max(phi_history) > 0.5:
    print("👍 PROMISING: Some evidence for self-referential consciousness")
else:
    print("🤔 INCONCLUSIVE: More investigation needed")
```

## Part 6: Multiple Experiments (Statistical Analysis)

### Reproducibility Check

Let's run multiple experiments to check if our results are reproducible:

```python
# Run multiple experiments for statistical analysis
num_experiments = 5
print(f"Running {num_experiments} experiments for statistical analysis...")

all_metrics = []
for exp in range(num_experiments):
    print(f"\nExperiment {exp + 1}/{num_experiments}")
    
    # Create fresh agent and environment
    test_agent = StrangeLoopAgent()
    test_env = create_environment_agent()
    
    # Run experiment
    test_results = run_experiment(test_agent, test_env, steps=50, verbose=False)
    all_metrics.append(test_results['metrics'])

# Aggregate and display results
print("\n" + "=" * 60)
print("AGGREGATED STATISTICS")
print("=" * 60)

metric_names = ['max_recursion_depth', 'avg_recursion_depth', 'final_phi', 'avg_phi', 'max_phi']

for metric in metric_names:
    values = [m[metric] for m in all_metrics]
    mean_val = np.mean(values)
    std_val = np.std(values)
    print(f"{metric:20s}: {mean_val:.3f} ± {std_val:.3f}")

# Check consistency of phi increases
phi_increases = [m['phi_increase'] for m in all_metrics]
positive_increases = sum(1 for inc in phi_increases if inc > 0)
print(f"\nConsistency: {positive_increases}/{num_experiments} experiments show Φ increase")

if positive_increases >= num_experiments * 0.8:  # 80% success rate
    print("✓ HIGH CONSISTENCY: Strange loops reliably increase consciousness!")
elif positive_increases >= num_experiments * 0.6:  # 60% success rate
    print("✓ MODERATE CONSISTENCY: Some reliability in results")
else:
    print("? LOW CONSISTENCY: Results may be unstable")
```
